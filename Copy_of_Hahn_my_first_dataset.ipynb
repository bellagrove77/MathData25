```python
# --- Step 3: Import libraries ---
from PIL import Image
import numpy as np
import os

# --- Step 4: Collect all image file names ---
img_files = [f for f in os.listdir() if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))]
print(f"Found {len(img_files)} images")

# --- Step 5: Resize all images to 256x256 and convert to NumPy array ---
target_size = (256, 256)
resized_images = []

for img_name in img_files:
    img = Image.open(img_name).convert("RGB")   # Ensure RGB
    img = img.resize(target_size)               # Resize to 256x256
    resized_images.append(np.array(img))

# Stack into a single NumPy array (N, 256, 256, 3)
images = np.stack(resized_images)

print("Final NumPy array shape:", images.shape)   # (num_images, 256, 256, 3)
print("Data type:", images.dtype)
```

```python
import matplotlib.pyplot as plt

# Show channels for the first 3 images
for idx in range(min(3, len(images))):
    img = images[idx]

    # Split channels
    R = img[:, :, 0]
    G = img[:, :, 1]
    B = img[:, :, 2]

    # Plot original + channels (color + grayscale)
    fig, axs = plt.subplots(2, 4, figsize=(16, 8))

    # Row 1 - Original + color-tinted channels
    axs[0,0].imshow(img)
    axs[0,0].set_title(f"Image {idx} - Original")
    axs[0,0].axis("off")

    axs[0,1].imshow(R, cmap="Reds")
    axs[0,1].set_title("Red (tinted)")
    axs[0,1].axis("off")

    axs[0,2].imshow(G, cmap="Greens")
    axs[0,2].set_title("Green (tinted)")
    axs[0,2].axis("off")

    axs[0,3].imshow(B, cmap="Blues")
    axs[0,3].set_title("Blue (tinted)")
    axs[0,3].axis("off")

    # Row 2 - grayscale versions
    axs[1,0].axis("off")  # empty spot for alignment

    axs[1,1].imshow(R, cmap="gray")
    axs[1,1].set_title("Red (grayscale)")
    axs[1,1].axis("off")

    axs[1,2].imshow(G, cmap="gray")
    axs[1,2].set_title("Green (grayscale)")
    axs[1,2].axis("off")

    axs[1,3].imshow(B, cmap="gray")
    axs[1,3].set_title("Blue (grayscale)")
    axs[1,3].axis("off")

    plt.show()
```
